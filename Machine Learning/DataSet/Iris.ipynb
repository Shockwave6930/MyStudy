## 2. データ操作

```python
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)   #こうすると、sklearn.utils.Bunch型のdataやtargetがDataFrameになる
#iris = load_iris(as_frame=False)   #こうすると、sklearn.utils.Bunch型のdataやtargetがndarray型になる
```

```python
type(iris)
```

- sklearn.utils.Bunchは簡単に言えば、dict型であれば_DictName_["_ KeyName _"]で値を取り出していたが、この型はdict型を継承して_DictName_._KeyName_のようにフィールド変数みたいに扱うことが可能になっている

> [Bunch クラス (sklearn.utils モジュール)
](https://biotech-lab.org/tech-info/python-ml/scikit-learn/library/sklearn-utils/bunch-class)

```python
iris.keys()   #sklearn.utils.Bunch型(ほぼdict型と見做して良い)のkey取得
```

```python
print(iris["DESCR"])   #["DESCR"], あるいは.DESCRにデータセットに関する詳細情報が格納されている
```
