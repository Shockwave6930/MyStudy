# 主成分分析:Principal Component Analysis

---

主成分分析は、相関のある多数の変数から相関のない少数で全体のばらつきを最もよく表す主成分と呼ばれる変数を合成する多変量解析の一手法。
データの次元を削減するために用いられる。

ここで、各データ(全て基準化済み)のベクトル$\boldsymbol{d_i}$と、主成分に並行な単位ベクトル$\boldsymbol{e}$を以下のようにおく。
$$
\begin{eqnarray}
\boldsymbol{d_i} &=& (x_i, y_i) \quad where \ i = 1, 2, ..., N\\
\boldsymbol{e} &=& (a, b)
\end{eqnarray}
$$
この時上図より各データの主成分方向の成分は$\boldsymbol{d_i}$と$\boldsymbol{e}$の内積$p_i$となる。 
$$
\begin{eqnarray}
p_i &=& \boldsymbol{d_i} \cdot \boldsymbol{e} \\
    &=& ax_i + by_i
\end{eqnarray}
$$
よって、この内積$p_i$のN=1,...,Nまでの分散が最大になれば良いことから、そのようなベクトル$\boldsymbol{e}$を求める。

分散$s^2$を次のように定義する。
$$
\begin{eqnarray}
s^2 = f(a, b) \equiv \displaystyle \sum_{i=1}^N \frac{p_i - \bar{p} }{N} \quad where \ \bar{p} = average(p_i)
\end{eqnarray}
$$
分散は$a, b$の関数であるので$f(a, b)$と書いた。これを展開すると、
$$
\begin{eqnarray}
f(a, b) &=& \displaystyle \sum_{i=1}^N \frac{((ax_i + by_i) - (a \bar{x} + b \bar{y}))^2}{N} \\
        &=& \displaystyle \sum_{i=1}^N \frac{(a(x_i - \bar{x}) + b (y_i - \bar{y}))^2}{N} \\
        &=& a^2 \displaystyle \sum_{i=1}^N \frac{(x_i - \bar{x})^2}{N} + 
        2ab \displaystyle \sum_{i=1}^N \frac{(x_i - \bar{x})(y_i - \bar{y})}{N} +
        b^2 \displaystyle \sum_{i=1}^N \frac{(y_i - \bar{y})^2}{N} \\
        &=& a^2s_x^2 + 2abs_{xy}^2 + b^2s_y^2
\end{eqnarray}
$$
となる。ここで$s_x^2$, $s_{xy}^2$, $s_y^2$はそれぞれ$x$, $y$の分散及び共分散である。
$$
s_x^2 = \displaystyle \sum_{i=1}^N \frac{(x_i - \bar{x})^2}{N} \quad
s_{xy}^2 = \displaystyle \sum_{i=1}^N \frac{(x_i - \bar{x})(y_i - \bar{y})}{N} \quad
s_y^2 = \displaystyle \sum_{i=1}^N \frac{(y_i - \bar{y})^2}{N}
$$

分散を既知の変数だけで書けたので、$\boldsymbol{e}=(a, b)$をラグランジュの未定乗数法を用いて求める。

> ラグランジュの未定乗数法　　  
ラグランジュの未定乗数法とは、ある拘束条件の下で関数を最大化/最小化させる変数を求めるための数学的手法である。  
ある関数$f(a, b)$が存在し、拘束条件として$g(a, b) = 0$が定められているとき、$L$を以下のように定義する。
$$
\begin{eqnarray}
L \equiv f(a, b) - \lambda g(a, b)
\end{eqnarray}
$$
この時次の関係を満たす$a, b$が目的の変数である。  
$$
\begin{eqnarray}
\frac{\partial L}{\partial a} = \frac{\partial L}{\partial b} = \frac{\partial L}{\partial \lambda} = 0 
\end{eqnarray}
$$
これは、各偏微分の値が0、つまり傾きが0の点を求めていることから極値が得られているのは自明である。  
ただし、それが最大値/最小値を与える点が得られているかは不明(局所解の可能性がある)なため、代入して適宜値を比較する必要がある。

今回の場合、拘束条件は$e$が単位ベクトルであることであるので、
$$
\begin{eqnarray}
|e| = \sqrt{a^2 + b^2} = 1
\end{eqnarray}
$$
から、$g(a, b)$は、
$$
g(a, b) = a^2 + b^2 - 1 = 0
$$
と書ける。
これを前述の引用内の式に代入すると、
$$
\begin{eqnarray}
  \left\{
    \begin{array}{1}
      \frac{\partial L}{\partial a} = as_x^2 + bs_{xy}^2 - a \lambda = 0 \\
      \frac{\partial L}{\partial b} = bs_y^2 + as_{xy}^2 - b \lambda = 0 \\
      \frac{\partial L}{\partial \lambda} = a^2 + b^2 - 1 = 0
    \end{array}
  \right.
\end{eqnarray}
$$
となり、これを解けばよい。

ここで第一式と第二式は行列形式で以下のように書ける。
$$
\begin{pmatrix}
s_x^2 & s_{xy}^2 \\
s_{xy}^2 & s_y^2
\end{pmatrix}
\left(
  \begin{array}{c}
    a \\
    b \\
  \end{array}
\right)
= \lambda
\left(
  \begin{array}{c}
    a \\
    b \\
  \end{array}
\right)
$$

左辺左側を分散共分散行列と呼び、データがあれば求まるため、これは$\lambda$と$(a, b)$を求める行列の固有値問題である。  
しかし、ここで求まった$(a, b)$はどの主成分なのかを特定する必要がある。  
ここで先に用語を定義する。
>射影するとデータの値の分散が最大になる主成分の軸のことを第一主成分(PC1)、同様に射影するとデータの値の分散がk番目に大きくなる軸のことを第k主成分(PCk)という。

先ほどの式の両辺に左から$(a, b)$を掛けてみる。  
$$  
\begin{eqnarray}
    (左辺) &=& (a, b) \lambda 
    \left(
    \begin{array}{c}
    a \\
    b \\
    \end{array}
    \right) \\
    &=& \lambda (a^2 + b^2) \\
    &=& \lambda \\ 
    (右辺) &=& (a, b)  
    \begin{pmatrix}
        s_x^2 & s_{xy}^2 \\
        s_{xy}^2 & s_y^2 
    \end{pmatrix}
    \left(
    \begin{array}{c}
    a \\
    b \\
    \end{array}
    \right) \\
&=& (a^2s_x^2 + abs_{xy}^2) + (abs_{xy}^2 + b^2s_y^2) \\
&=& a^2s_x^2 + 2abs_{xy}^2 + b^2s_y^2 \\
&=& s^2
\end{eqnarray}
$$

よって、
$$\lambda = s^2 = f(a, b)$$
であることが分かる。  
このことから、最も大きい固有値に対応する固有ベクトルが第一主成分の単位ベクトルであり、二番目に大きい固有値に対応する固有ベクトルが第二主成分の単位ベクトル、...であることが分かる。  
また、共分散分散行列は実対称行列(実エルミート行列)であるため、相異なる固有値に対応する固有ベクトルは直交するため、PC1, PC2, ...は全て直交する。  
この方法で得られた主成分に次元を落とすことによって情報をなるべく失うことなく次元縮減が可能である。

ここからは、主成分分析の結果を見る際に扱う各種指標について説明する
- 主成分得点  
$
y_i = h_{i1}x_1 + h_{i2}x_2 + \ldots + h_{ip}x_p
$  
個々のデータから求めた主成分の値のこと
- 標準偏差  
該当する主成分が持つ情報量を表す
この標準偏差は上記の式 $\lambda = s^2$ から固有値$\lambda$の平方根となっている

- 寄与率
$$
\begin{eqnarray}
(寄与率) = \frac{l_i}{（総分散）} \qquad where \ (総分散) = \sum_{i=1}^N l_i, \quad l_i = (第i固有値)
\end{eqnarray}
$$
全情報量のうち、該当する主成分が占める情報量の割合
- 累積寄与率 
$$
\begin{eqnarray}
(累積寄与率) = \frac{l_1 + l_2 + \ldots + l_k}{（総分散）}
\end{eqnarray}
$$
第一主成分から第k主成分までの主成分が占める情報量の割合  
次元縮減によって失う情報量を測ることができる
- 因子負荷量  
$$
Cor(x_i, y_j) = \sqrt{l_j} h_{ij} \quad where \ x_i = (i番目の変数), \quad y_j = (第j主成分), \quad h_{ij} = (第j主成分のi番目の変数の係数)
$$
主成分と各変数の値との相関係数  
その主成分にどれだけその変数が寄与しているかの指標になる  
多変量解析の分野で出てくる言葉でちゃんとは分かってないから調べておく

## Ex)3種類13次元のワインのデータを二次元に縮減して種類ごとに分けられるか試してみる

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Materials/Data/wine.txt", sep="\t", index_col=0)
df
```

```python
from pandas import plotting 
plotting.scatter_matrix(df.iloc[:, 1:], figsize=(16, 16), c=list(df.iloc[:, 0]), alpha=0.5)
plt.show()
```

```python
#基準化
dfs = df.iloc[:, 1:].apply(lambda x: (x-x.mean())/x.std(), axis=0)
dfs.head()
```

```python
#主成分分析の実行
pca = PCA()
pca.fit(dfs)
# データを主成分空間に写像
feature = pca.transform(dfs)
```

```python
# 主成分得点
pd.DataFrame(feature, columns=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 第一主成分と第二主成分でプロットする
plt.figure(figsize=(6, 6))
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, c=list(df.iloc[:, 0]))
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from pandas import plotting 
plotting.scatter_matrix(pd.DataFrame(feature, 
                        columns=["PC{}".format(x + 1) for x in range(len(dfs.columns))]), 
                        figsize=(16, 16), c=list(df.iloc[:, 0]), alpha=0.5) 
plt.show()
```

```python
# 寄与率
pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 累積寄与率を図示する
import matplotlib.ticker as ticker
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))
plt.plot([0] + list( np.cumsum(pca.explained_variance_ratio_)), "-o")
plt.xlabel("Number of principal components")
plt.ylabel("Cumulative contribution rate")
plt.grid()
plt.show()
```

```python
# PCA の固有値
pd.DataFrame(pca.explained_variance_, index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# PCA の固有ベクトル
pd.DataFrame(pca.components_, columns=df.columns[1:], index=["PC{}".format(x + 1) for x in range(len(dfs.columns))])
```

```python
# 第一主成分と第二主成分における観測変数の寄与度をプロットする
plt.figure(figsize=(6, 6))
for x, y, name in zip(pca.components_[0], pca.components_[1], df.columns[1:]):
    plt.text(x, y, name)
plt.scatter(pca.components_[0], pca.components_[1], alpha=0.8)
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
